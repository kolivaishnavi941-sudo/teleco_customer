```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
# preprocessing part
# in this part wer will apply standardscaler

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# creating model
model=LogisticRegression()


x_train = np.nan_to_num(x_train, nan=1.0)

# traning model
model.fit(x_train,y_train)
# here the ML actually learns patern
```

```python
# prediction
# predict train and test
y_train_pred=model.predict(x_train)

# Handle NaN values in x_test before prediction, similar to x_train
x_test = np.nan_to_num(x_test, nan=1.0)
y_test_pred=model.predict(x_test)
```

```python
x_trained = pd.DataFrame(x_train,columns=x.columns)
x_trained.head()
```

```python
#accuracy check
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print(train_acc,test_acc)
# Accuracy alone is not enough for classification
# if 90 died, 10 survived but the model says everyone died
# the model will be useless
# here is why new are using confusion matrix
```

```python
# properfit-high train and high test with a small gap
# overfit-big gap(train>>test)
# underfit-both train and test perfrom poorly
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_test_pred)
cm
# [tn fp]
# [fn tp]
# tn= true negetive - prdicted died and actuallyt died
# fp= flase negetive -predicted survived and actually died
# fn flase negetive- predicted died and actually survived
# tp = true positive- predcted survived and actually survived

```

```python
# confusion matrix is to denote where the model got confusied
# where the model made dangeropus mistake
# a confusion matrix is a table used in classification that
# compares actual and predicted  classes, showing the
# number of correct and incorrect predictions.it providesd much
# more insight than accuracy alone and is the foundation for matrics
# like prediction,recall and f1-score
```

```python
# visual confusion matrix
sns.heatmap(cm,annot=True,cmap='Blues')
plt.title('confusion matrix')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
#
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_pred))
# this gives all metrics autometically
```

```python
# prediction
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)

from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_pred_train)
test_acc=accuracy_score(y_test,y_pred_test)
print(train_acc,test_acc)
```
