```python
#  columns where 0 is not valid medical value
# glucose,bloodpressure,shinthickness,insulin,bmi
invalid_cols=['glucose','bloodpressure','skinthickness','insulin','bmi']
# rplace 0 with nan
df[invalid_cols]= df[invalid_cols].replace(0,np.nan)
```

```python
# impute mi9ssing values
for col in invalid_cols:
  df[col].fillna(df[col].median(),inplace=True)
  # medium represents a typical patient not too high not too low
  # not in fulence by outliers
```

```python
df.isnull().sum()
```

```python
#  feature and target separetion
# feature
x=df.drop('outcome',axis=1)
# target
y=df['outcome']
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100,stratify=y)
```

```python
# feature scaling preprocessing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train[:10]
```

```python
# here
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#  predictions
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,y_pred_train)
```

```python
# classification report
from sklearn.metrics import classification_report
cr=(classification_report(y_train,y_pred_train))
print(cr)
```
